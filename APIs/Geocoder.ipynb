# Geocoder API

```python
import requests as rq
import json
from pprint import pprint
```

## Finn steder som matcher "Oslo"

```python
r = rq.get("https://api.entur.io/geocoder/v1/autocomplete?text=Oslo&size=20&lang=no")
pprint(json.loads(r.content))
```

## Finn steder innenfor en "bounding box"

```python
url = "https://api.entur.io/geocoder/v1/reverse?point.lat=60.1&point.lon=10.04&boundary.circle.radius=1&size=10&layers=address%2Clocality"
response = json.loads(rq.get(url).content)
print(response.keys())
pprint(response["features"])
```

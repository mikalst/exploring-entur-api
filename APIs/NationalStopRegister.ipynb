# National Stop Register API

National Stop Register (NSR) is the master database for public transport stops in Norway and is primarily used to store and redistribute detailed information regarding the infrastructure of a stop place.

```python
import requests as rq
import json
from pprint import pprint
```

## Hent et "sted"??

```python
r = rq.get("https://api.entur.io/stop-places/v1/read/topographic-places?count=1")
pprint(json.loads(r.content))
```

## Hent en parkering

```python
r = rq.get("https://api.entur.io/stop-places/v1/read/parkings?count=1")
pprint(json.loads(r.content))
```
